```python
df_corr = df.corr()
fig, ax = plt.subplots(figsize=(15, 5))
mask = np.triu(np.ones_like(df_corr, dtype=np.bool_))
mask = mask[1:, :-1]
corr = df_corr.iloc[1:,:-1].copy()
sns.heatmap(corr,linewidths=.5, mask=mask, annot=True, fmt=".1f", cmap='Blues',
           vmin=-1, vmax=1, cbar_kws={"shrink": .8})
plt.yticks(rotation=0)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
mask = np.triu(np.ones_like(df_corr, dtype=np.bool_))
mask = mask[1:, :-1]
corr = df_corr.iloc[1:,:-1].copy()
cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", 
           linewidths=5, cmap=cmap, vmin=-1, vmax=1, 
            cbar_kws={"shrink": .8})
yticks = [i.upper() for i in corr.index]
xticks = [i.upper() for i in corr.columns]
plt.yticks(plt.yticks()[0], labels=yticks, rotation=45)
plt.xticks(plt.xticks()[0], labels=xticks)
# title
title = 'POKEMON\nWEEDLES CAVE\n'
plt.title(title, loc='center', fontsize=18)
plt.show()
```
